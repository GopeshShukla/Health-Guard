```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
heart = pd.read_csv('heart.csv')
```

```python
# separating data into dependent and independent.

x = heart.iloc[:,0:13]
x
```

```python
y = heart.iloc[:,-1]
y
```

```python
# splitting data into training and testing data.

x_train,x_test,y_train,y_test = train_test_split(x, y, random_state = 5, test_size=0.2)
```

```python
# traing model

classifier1 = svm.SVC(kernel='linear')
classifier1.fit(x_train,y_train)
```

```python
y_pred = classifier1.predict(x_test)
accuracy = accuracy_score(y_pred,y_test)
accuracy
```

```python
pred = classifier1.predict([[63,1,3,145,233,1,0,150,0,2.3,0,0,1]])
if pred == 1:
    print(' Patient is suffering from Heart Problem. ')

else:
    print( 'Patient is not suffering from Heart Problem.' )
```

```python
# dumping this ML model as pkl file.

import pickle
pickle.dump(classifier1,open('heart.pkl','wb'))
```

```python
# opening pkl file.

model = pickle.load(open('heart.pkl','rb'))
```
